```python
!pip install qmcpy --quiet
import qmcpy
import numpy as np
from matplotlib import pyplot as plt
```

## 実験目的

ハルトン列の一様性を調べ、置換の有効性を確認する


```python
# 正方形上の散布図向けの設定にする
def set_scatter_plot(axes):
    axes.set_xticks([0,1])
    axes.set_yticks([0,1])
    axes.set_xlim([0,1])
    axes.set_ylim([0,1])
    axes.set_aspect('equal')

def show_Halton_plots(dim1,dim2,number_of_points,generators):
    k = len(generators)
    fig, axes = plt.subplots(ncols=k)
    for i,generator in enumerate(generators):
        points = generator.gen_samples(number_of_points)
        axes[i].scatter(points[:,dim1], points[:,dim2],color="black",s=5)
        set_scatter_plot(axes[i])

    plt.savefig(f"Halton{dim1}_{dim2}.pdf",format="pdf",bbox_inches='tight')
    plt.show()
```


## 実験

- $1,2$ 次元目 ($p=2,3$)のハルトン列を図示する
- $17,18$ 次元目($p=59,61$)のハルトン列を図示する
- $18,19$ 次元目($p=61,67$)のハルトン列を図示する


```python
NOrand = qmcpy.Halton(25,randomize=False,generalize=False)
DOrand = qmcpy.Halton(25,randomize=True)

show_Halton_plots(0,1,1000,[NOrand,DOrand])
show_Halton_plots(16,17,1000,[NOrand,DOrand])
show_Halton_plots(17,18,1000,[NOrand,DOrand])
```

## 結果

隣接した次元の射影を見ると、均等性が悪い。
とくに対応する素数が双子素数の場合、非常に均等性が悪くなる。
